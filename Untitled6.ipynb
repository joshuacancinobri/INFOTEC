```python
def validar_datos(nombre, edad, correo):
    # Inclusión de dos mensajes de depuración
    print("\nValidando datos...")
    print(f"Datos recibidos -> Nombre: {nombre}, Edad: {edad}, Correo: {correo}")
    try:
        #Validar que nombre sea una cadena no vacía
        if not isinstance(nombre, str):
            raise TypeError("El nombre debe ser una cadena de texto.")
        if not nombre:
            raise ValueError("El nombre no puede estar vacío.")

        #Validar que edad sea un entero mayor que cero
        if not isinstance(edad, int):
            raise TypeError("La edad debe ser un número entero.")
        if edad <= 0:
            raise ValueError("La edad debe ser un número entero mayor que cero.")

        #Validar que correo contenga un '@'
        if not isinstance(correo, str):
            raise TypeError("El correo debe ser una cadena de texto.")
        if '@' not in correo:
            raise ValueError("El correo debe contener un símbolo @.")

        #ZeroDivisionError con 1
        if edad == 1:
            raise ZeroDivisionError("La edad no puede ser 1.")

        # Si todo está bien
        print("Datos registrados correctamente.")

    # TypeError
    except TypeError as e:
        print(f"Error de tipo: {e}")
        print("Datos no registrados.")

    # ValueError
    except ValueError as e:
        print(f"Error de valor: {e}")
        print("Datos no registrados.")

    # ZeroDivisionError
    except ZeroDivisionError as e:
        print(f"Error de lógica (división): {e}")
        print("Datos no registrados.")

    finally:
        print("Registro finalizado.")

def registrar_usuario():
    nombre = input("Introduce el nombre: ")
    correo = input("Introduce el correo: ")
    edad_str = input("Introduce la edad: ")

    try:
        edad = int(edad_str)
    except ValueError:
        edad = edad_str
    validar_datos(nombre, edad, correo)

registrar_usuario()
```
